```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# Load box score data containing all players' and teams' info for all regular season NBA games from 2012-2018

player = pd.read_csv(r'C:\Users\JohnSteph\Documents\thinkful\nba-enhanced-stats\2012-18_playerBoxScore.csv')
team = pd.read_csv(r'C:\Users\JohnSteph\Documents\thinkful\nba-enhanced-stats\2012-18_teamBoxScore.csv')
```

```python
# Start with something obvious: field goal attempts (continuous) separated by whether the player is a starter or reserve (categorical)

sns.catplot(x="playStat", y="playFGA", data=player);
```

```python
# Compare how many fouls are called on home teams vs. road teams

sns.catplot(x="teamLoc", y="teamPF", kind="boxen", data=team);
```

```python
# Use a violin plot to compare the FGA data from above combined with home/road info

sns.catplot(x="teamLoc", y="playFGA", hue="playStat", kind="violin", data=player);
```

```python
# Plot points scored vs. FG attempts, separated by home/road

sns.lmplot(y='teamPTS', x='teamFGA', hue='teamLoc', data=team, fit_reg=True, scatter_kws={'alpha':0.4})
```
